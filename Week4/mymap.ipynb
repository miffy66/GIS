```python
# import folium, a leaflet python library
import folium

# create a list of lat/lon pair for 麗澤大学近くのクラフトビール屋さん
reitaku = [35.8330, 139.9550]
tsumugi = [35.8501, 139.9634]
kashiwa = [35.8317, 139.9713]
life = [35.8662, 139.9685]
glover = [35.8606, 139.9732]
cluster = [35.8633, 139.9706]
house = [35.8620, 139.9723]

# make the map
m = folium.Map(location=reitaku and tsumugi and kashiwa and life and glover and cluster and house , zoom_start=13)

# add a marker
folium.Marker(reitaku, popup="Reitaku University!", tooltip="Reitaku University").add_to(m)
folium.Marker(tsumugi, popup="4.8", tooltip="2階のクラフトビール屋つむぎ"). add_to(m)
folium.Marker(kashiwa, popup="4.3", tooltip="柏ビール(こまいぬブルワリー タップルーム)"). add_to(m)
folium.Marker(life, popup="3.9", tooltip="樽生クラフトビール15TAPと自家製ソーセージ『THE LIFE』"). add_to(m)
folium.Marker(glover, popup="4.2", tooltip="グラバー 柏店"). add_to(m)
folium.Marker(cluster, popup="4.2", tooltip="地産地消 Cafe & Bar Cluster"). add_to(m)
folium.Marker(house, popup="4.5", tooltip="Craft House Kashiwa"). add_to(m)

# show the map
m
```

```python
m.save("map.html")
```
